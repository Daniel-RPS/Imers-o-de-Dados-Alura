## Desafio 4 - Verificar se quem zerou a prova foi eliminado ou estava ausente

##### 0 - Faltou à prova
##### 1 - Presente na prova
##### 2 - Eliminado na prova


* 72,7% dos candidatos fizeram todas as provas. 
* 22,76% dos candidatos faltaram à todas as provas. 
* 4,45% dos candidatos podem ter feito 2 provas em um final de semana e faltado no outro final de semana
* 0,05% dos candidatos foram eliminados em um fim de semana, mas conseguiu fazer as outras 2 provas em outro fim de semana.

```python
presenca = ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']
dados['PRESENCA_TOTAL'] = dados[presenca].sum(axis=1)
presenca.append("PRESENCA_TOTAL")
dados[presenca]
```

* Aqui, podemos chegar à conclusão de que quem fez todas as prova, terá a PRESENCA_TOTAL igual a 4, e quem não fez, igual 2. 
* Isso porque foram realizadas 4 provas, 2 em cada fim de semana, então quem fez a prova de matemática, acabou fazendo a prova de ciências naturais, por exemplo, e quem não fez a prova de ciências humanas, não fez a prova de linguagens e códigos. 
* Ainda temos o valor 6, que significa que o candidato foi eliminado em um fim de semana, mas conseguiu fazer as outras 2 provas em outro fim de semana.

```python
# Dados sobre a prova de ciências naturais (Mas que serve para a prova de Matemática, pois ambas foram realizadas no mesmo dia)
cn_mt = dados['TP_PRESENCA_CN'].value_counts()
cn_mt
```

```python
plt.figure(figsize=(7,7))
plt.pie(cn_mt, labels=['1', '0', '2'], autopct='%1.2f%%', shadow=True);
```

#### 72,8% dos candidatos fizeram as provas. 27% faltaram às provas e apenas 0,04% dos candidatos foram eliminados.

```python
# Dados sobre a prova de ciências humanas (Mas que serve para a prova de Linguagens e Códigos, pois ambas foram realizadas no mesmo dia)
ch_lc = dados['TP_PRESENCA_CH'].value_counts()
ch_lc
```

```python
plt.figure(figsize=(7,7))
plt.pie(ch_lc, labels=['1', '0', '2'], autopct='%1.2f%%', shadow=True);
```

#### 77% dos candidatos fizeram as provas. 23% faltaram às provas e apenas 0,1% dos candidatos foram eliminados.

```python
# Total em relação às outras provas
p_total = dados['PRESENCA_TOTAL'].value_counts()
p_total
```

```python
plt.figure(figsize=(7,7))
plt.pie(p_total, labels=['4', '0', '2', '6'], autopct='%1.2f%%', shadow=True);
```

#### 72,7% dos candidatos fizeram todas as provas. 22,76% dos candidatos faltaram à todas as provas. 4,45% dos candidatos podem ter feito 2 provas em um final de semana e faltado no outro final de semana, e 0,05% dos candidatos foram eliminados em um fim de semana, mas conseguiu fazer as outras 2 provas em outro fim de semana.

## Desafio 6 - Verificar a proporção dos participantes de rendas mais altas e mais baixas como treineiro e não treineiro

* Entre os treineiros, a maioria se enquadra na classe B, com renda de até 998,00 reais. Entre os não-treineiros, a maioria se enquadra na classe C, com renda de 998,01 até 1.497,00 reais.

```python
#Dados dos treineiros
treineiros = pd.DataFrame(dados.query("IN_TREINEIRO == 1")["Q006"].value_counts(normalize=True))
treineiros = treineiros["Q006"]*100
treineiros
```

## Desafio 7 - Fazer o mesmo bloxplot olhando para a Questão 25 (tem internet ou não) e fazer uma reflexão sobre o assunto e o contexto de pandemia

##### Na sua residência tem acesso à Internet?
##### A - Não
##### B - Sim

* Olhando o boxplot pelo contexto da pandemia, é possível ver que, quem possui internet terá notas melhores em relação aos que não têm. A discrepância é maior nas classes mais altas, pois a maioria possui acesso à internet.

```python
dados_sem_zero = dados.query('NU_NOTA_TOTAL != 0')
dados_sem_zero.head()
```

```python
renda_order = dados['Q006'].unique() 
renda_order.sort() #Ordenando a renda

plt.figure(figsize=(15,6))
plt.title('Total das notas pela renda')
sns.boxplot(x='Q006', y='NU_NOTA_TOTAL', data=dados_sem_zero, hue= "Q025",
            order= renda_order );
```
