## Desafio 1 - Proporção de inscritos por idade

* A maioria dos candidatos possuem 18 anos (16.84%)

```python
proporcao = dados["NU_IDADE"].value_counts(normalize=True)
proporcao
```

```python
proporcao*100
```

## Desafio 2 - Descobrir de quais estados são os inscritos de 13 anos

* São Paulo, Amapá, Bahia e Mato Grosso

```python
dados.query('NU_IDADE == 13')['SG_UF_RESIDENCIA'].value_counts()
```

## Desafio 3 - Adicionar título no gráfico

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Gráfico em pizza de idades', fontsize=19)
graf = dados["NU_IDADE"].value_counts(normalize=True)
graf.plot.pie(figsize=(15,10));
```

## Desafio 4 - Plotar os histogramas das idades dos treineiros e não treineiros

* Os não treineiros estão em quantidade superior em relação aos treineiros. Significa que a maioria dos candidatos foi para tentar uma vaga em alguma instituição de ensino superior, e não para apenas treinar.

```python
plt.figure(figsize=(12,6))
dados.query('IN_TREINEIRO == 0')['NU_IDADE'].hist(bins=15, label='Não treineiros')
dados.query('IN_TREINEIRO == 1')['NU_IDADE'].hist(bins=15, label='Treineiros')
plt.title('Histograma de Treineiros e Não Treineiros')
plt.legend();
```

## Desafio 5 - Comparar as distribuições das provas em inglês e espanhol

* A maioria dos candidatos fez a prova de espanhol. A nota das maioria dos candidatos ficou entre 400 e 600.

* TP_LINGUA
    * 0: Inglês
    * 1: Espanhol

```python
dados['TP_LINGUA'].value_counts()
```

```python
plt.figure(figsize=(12,6))
dados.query('TP_LINGUA == 0')['NU_NOTA_LC'].hist(bins=15,label='Inglês')
dados.query('TP_LINGUA == 1')['NU_NOTA_LC'].hist(bins=15,label='Espanhol')
plt.title('Histograma de Línguas Estrangeiras',fontsize=19)
plt.legend();
```

## Desafio 6 - Gerar novas visualizações com matplotlib e pandas

* Em média, as notas de quem tem acesso à internet na residência é superior em relação aos candidatos que não possuem internet em sua residência.

* Q025: Na sua residência tem acesso à Internet?
    * A: Não
    * B: Sim

```python
media_internet = dados[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','Q025']].groupby('Q025').mean().reset_index()
media_internet
```

```python
media_internet.plot(figsize=(12,6));
```
