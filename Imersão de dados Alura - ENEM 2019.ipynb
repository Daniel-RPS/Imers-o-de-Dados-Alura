* stat: Aggregate statistic to compute in each bin:


* count shows the number of observations

* frequency shows the number of observations divided by the bin width

* density normalizes counts so that the area of the histogram is 1

* probability normalizes counts so that the sum of the bar heights is 1

## Aula 4

```python
provas
```

```python
provas_entrada = ['NU_NOTA_CH','NU_NOTA_CN','NU_NOTA_LC','NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'

dados_sem_zero = dados_sem_zero[provas].dropna()

notas_entrada = dados_sem_zero[provas_entrada]
notas_saida = dados_sem_zero[prova_saida]
```

```python
#f(x) = y
x = notas_entrada
y = notas_saida
```

```python
#Separando dados de treino e dados de teste

from sklearn.model_selection import train_test_split

seed = 4321

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, random_state=seed)
```

```python
x_treino.head()
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
# Criando o modelo
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state=seed)
modelo.fit(x_treino,y_treino)
```

```python
#Predição dos valores a serem avaliados (previsão)
predicoes_mt = modelo.predict(x_teste)
predicoes_mt
```

```python
#Valores reais
y_teste[:5]
```

* Valores reais//Previsão

* 459.7 // 420.31950108
* 617.2 // 478.71700082
* 520.0 // 385.7069269
         ...


```python
plt.figure(figsize=(7,7))
sns.scatterplot(x=predicoes_mt,y=y_teste,)
plt.xlim((-50,1050))
plt.ylim((-50,1050));
```

```python
# O erro do modelo é grande. Quem tirou, por exemplo, 400, acabou recebendo um valor negativo pelo modelo. 
# Quem tirou acima de 700, acabou sendo beneficiado pelo modelo.

plt.figure(figsize=(7,7))
sns.scatterplot(x=y_teste,y=y_teste - predicoes_mt);
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x=y_teste,y=x_teste.mean(axis=1));
```

```python
resultados = pd.DataFrame()
resultados['Real'] = y_teste
resultados['Previsão'] = predicoes_mt
resultados['Diferença'] = resultados['Real'] - resultados['Previsão']
resultados['quadrado_diferença'] = (resultados['Real'] - resultados['Previsão'])**2
```

```python
resultados
```

```python
#Media
resultados['quadrado_diferença'].mean()
```

```python
#Raiz quadrada
resultados['quadrado_diferença'].mean()**(1/2)
```

#### Comparando com outros modelos (como o Dummy)

```python
from sklearn.dummy import DummyRegressor

modelo_dummy = DummyRegressor()
modelo_dummy.fit(x_treino, y_treino)
dummy_predicoes = modelo_dummy.predict(x_teste)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_teste, dummy_predicoes)
```

```python
#Resultado é o mesmo de resultados['quadrado_diferença'].mean()
mean_squared_error(y_teste, predicoes_mt)
```
